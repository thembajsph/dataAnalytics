```python
import decimal


def homeLoanApplication():
    # granted = True
    # rejected = True
    print("Hi there I'm Andile, your home loan digital assistance :D...")
    gross_salary = float(input("Enter your gross salary: "))
    credit_score = float(input("Now enter your credit score(0-1200): "))
    purchase_price = float(input("What is the purchase price of the property: "))
    # print("Thank you for your home loan application.")
    afford = float(0.3 * gross_salary)
    deposit = 0
    bond_amount = purchase_price - deposit
    monthly_instalment = bond_amount * 0.00785
    has_good_salary = afford > monthly_instalment
    # max_instalments at 7.5% interest rate for 20 years mortgage
    answer = float((bond_amount * (1 / 160)))
    max_instalment = float(answer / 0.77582582)
    good_record = monthly_instalment < max_instalment

    #  1. The instalment of a home loan should be no more than 30% of a customer’s gross salary.
    #     2. The home loan percentage granted is determined by the credit score as below:

    #     • 800 and above = 110% loan.
    if credit_score == 800 and has_good_salary and good_record:

        #   Percentage_Home_Loan_granted = 110%
        print("Loan application status : granted")
        # print(F"Maximum Instalment Amount : ${max_instalment}")
        print(F"Maximum Instalment Amount : ${ afford}")
        print("Percentage Home Loan granted : 110%")
        deposit = purchase_price * 0
        print(F"Deposit required : ${deposit}")
        monthly_instalment = (bond_amount + (bond_amount * 0.1)) * 0.00785
        print(F"Monthly Instalment : ${monthly_instalment}")
        print("Press Any Key to continue... ")
    #

    #     • 750-799 = 100% loan.
    elif has_good_salary and 750 <= credit_score <= 799 and good_record:

        # Percentage_Home_Loan_granted = 100%
        print("Loan application status : granted")
        # print(F"Maximum Instalment Amount : ${max_instalment}")
        print(F"Maximum Instalment Amount : ${ afford}")
        print(F"Percentage Home Loan granted : 100%")
        deposit = purchase_price * 0
        print(F"Deposit required : ${deposit}")
        print(F"Monthly Instalment : ${monthly_instalment}")
        print("Press Any Key to continue... ")




    #     • 700-749 = 95% loan.
    elif has_good_salary and 700 <= credit_score <= 749 and good_record:

        # Percentage_Home_Loan_granted = 95%
        print("Loan application status : granted")
        # print(F"Maximum Instalment Amount : ${max_instalment}")
        print(F"Maximum Instalment Amount : ${ afford}")
        print(F"Percentage Home Loan granted : 95%")
        deposit = purchase_price * 0.05
        print(F"Deposit required : ${deposit}")
        print(F"Monthly Instalment : ${monthly_instalment}")
        print("Press Any Key to continue... ")

    #     • 650-699 = 90% loan.
    elif has_good_salary and 650 <= credit_score <= 699 and good_record:
        # Percentage_Home_Loan_granted = 90%
        print("Loan application status : granted")
        # print(F"Maximum Instalment Amount : ${max_instalment}")
        print(F"Maximum Instalment Amount : ${ afford}")
        print(F"Percentage Home Loan granted : 90%")
        deposit = purchase_price * 0.1
        print(F"Deposit required : ${deposit}")
        print(F"Monthly Instalment : ${monthly_instalment}")
        print("Press Any Key to continue... ")

    #     • 5.  600-649 = 85% loan.
    elif has_good_salary and 600 <= credit_score <= 649 and good_record:

        # Percentage_Home_Loan_granted = 85%
        print("Loan application status : granted")
        # print(F"Maximum Instalment Amount : ${max_instalment}")
        print(F"Maximum Instalment Amount : ${ afford}")
        print(F"Percentage Home Loan granted : 110%")
        deposit = purchase_price * 0.15
        print(F"Deposit required : ${deposit}")
        print(F"Monthly Instalment : ${monthly_instalment}")
        print("Press Any Key to continue... ")


    #     • 6.  550-599 = 80% loan.
    elif has_good_salary and 550 <= credit_score <= 599 and good_record:

        # Percentage_Home_Loan_granted = 80%
        print("Loan application status : Granted")
        # print(F"Maximum Instalment Amount : ${max_instalment}")
        print(F"Maximum Instalment Amount : ${ afford}")
        print(F"Percentage Home Loan granted : 80%")
        deposit = purchase_price * 0.2
        print(F"Deposit required : ${deposit}")
        print(F"Monthly Instalment : ${monthly_instalment}")
        print("Press Any Key to continue... ")


    #     • 7. 549 and below = rejected.

    else:
        print("Loan application status : rejected")
```
