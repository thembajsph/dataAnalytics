```python
#(3) Which 3 Territories have grown their Revenue the most year on year ?
```

```python
df3Territories = pd.merge(dfSales, dfDoctors  , on= "DoctorID")
```

```python
df3TerritoriesFinal = pd.merge(df3Territories, dfTerritory  , on= "Territory")
```

```python
df3TerritoriesFinal.head(5)
```

```python
# sorted_df = (( df3TerritoriesFinal['Year'] == 2021) & (df3TerritoriesFinal['Revenue']) > (df3TerritoriesFinal['Year'] == 2020) & (df3TerritoriesFinal['Revenue'])) 
```

```python
totalRevenuesNew = df3TerritoriesFinal.groupby(["Territory", 'Year'],as_index=False,)['Revenue'].sum() 
```

```python
totalRevenuesNew.head(5)
```

```python
output = totalRevenuesNew.assign(annual_change=totalRevenuesNew.groupby("Territory")['Revenue'].apply(lambda x:x.diff().fillna(x)))
```

```python
output.head(5)
```

```python
sorted_dfTopRevenueProduct = (output['Year'] == 2021)
```

```python
sorted_dfNewPro = output[ sorted_dfTopRevenueProduct ]
# print(sorted_dfNewTeritorries.shape)
sorted_dfNewPro.head(5)
```

```python
sorted_dfTopRevenueProduct = sorted_dfNewPro.sort_values(["annual_change"] , ascending=False)
```

```python
sorted_dfTopRevenueProduct.head(3)
```
